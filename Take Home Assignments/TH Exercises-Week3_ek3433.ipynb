```python
#1 ⁠Filter the data to include only weekdays (Monday to Friday) and plot a line graph showing the pedestrian counts for each day of the week.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

#I was getting a few Future warning messages that got displayed in the code so I wanted to add this line to avoid cluttering
warnings.simplefilter(action='ignore', category=FutureWarning)

# Read the dataset

url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"

df = pd.read_csv(url)

#Display the initial df
display(df)

# Inspect column names to check which one has the dates
print(df.columns)

# Based on investigation: Convert 'hour_beginning' to datetime
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])

#Arranging data in chronological order first and then proceeding to forward/back fill
df = df.sort_values(by='hour_beginning') 
#drop=True discards the current index, inplace=True applies changes to the df
df.reset_index(drop=True, inplace=True)

df['temperature'] = df['temperature'].fillna(method="ffill")
df['precipitation'] = df['precipitation'].fillna(method="ffill")
df['weather_summary'] = df['weather_summary'].fillna(method="ffill")

# Extract the day of the week (0=Monday, 6=Sunday)
df['hour'] = df['hour_beginning'].dt.hour
df['month'] = df['hour_beginning'].dt.month
df['year'] = df['hour_beginning'].dt.year
df['date'] = df['hour_beginning'].dt.date
df['day_name'] = df['hour_beginning'].dt.day_name()
df['Weekday'] = df['hour_beginning'].dt.dayofweek

#Display new df with the added column
display(df)

# Filter only weekdays (Monday=0 to Friday=4)
df_weekdays = df[df['Weekday'] < 5]
display(df_weekdays)

# Aggregate pedestrian counts by weekday using groupby
weekday_counts = df_weekdays.groupby('Weekday')['Pedestrians'].sum()

# Plot the pedestrian count per weekday
plt.figure(figsize=(10, 5))
plt.plot(weekday_counts.index, weekday_counts.values, marker='s')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])
plt.xlabel('Day of the Week')
plt.ylabel('Total Pedestrian Count')
plt.title('Pedestrian Counts for Each Weekday')
plt.grid(True)
plt.show()
```

```python
#2 ⁠Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. 
# Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) 
# between weather patterns and pedestrian counts for the selected year.
# This question requires you to show the relationship between a numerical feature(Pedestrians) and a non-numerical feature(Weather Summary). 
# In such instances we use Encoding. Each weather condition can be encoded as numbers( 0,1,2..). This technique is called One-hot encoding.
# Correlation matrices may not always be the most suitable visualization method for relationships involving categorical data points, 
# nonetheless this was given as a question to help you understand the concept better.

# Filter for 2019 data abd sort by weather_summary
df_2019 = df[df['year'] == 2019]
df_2019 = df.sort_values(by='weather_summary')
#drop=True discards the current index, inplace=True applies changes to the df
df_2019.reset_index(drop=True, inplace=True) 

# Get unique weather conditions
weather_types = df_2019['weather_summary'].unique()
weather_types.sort()

# Print the manual weather encoding for reference
for type in range(len(weather_types)):
    print("Weather Encoding Mapping", type,":", weather_types[type])

#Create column to place encoded values
df_2019['weather_encoded']=0

# Encode the unique values manually, sorted alphabetically
for i in range(len(df_2019)):
    for j in range(len(weather_types)):
        if df_2019.loc[i,'weather_summary']== weather_types[j]:
            df_2019.loc[i,'weather_encoded']=j

# Compute correlation matrix (Pedestrians vs. Encoded Weather)
correlation_matrix = df_2019[['Pedestrians', 'weather_encoded']].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Weather Conditions and Pedestrian Counts (2019)")
plt.show()


```

```python
#3.⁠ ⁠Implement a custom function to categorize time of day into morning, afternoon, evening, and night, and create a new column in the DataFrame to store these categories. 
# Use this new column to analyze pedestrian activity patterns throughout the day.-Students can also show plots analyzing activity.

# Custom function to categorize time of day
# The times corresponding to each period was assigned based on how I (ek3433) view when morning, afternoon, evening, and night times start
def categorize_time(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

# Apply function to create new column
df['time_of_day'] = df['hour'].apply(categorize_time)

#Get average counts for the categorized times of day and display them
time_of_day_counts = df.groupby('time_of_day')['Pedestrians'].mean()
display(round(time_of_day_counts))

#Get total counts for the categorized times of day and display them
time_of_day_total = df.groupby('time_of_day')['Pedestrians'].sum()
display(round(time_of_day_total))

# Display first 25 rows to check
display(df[['hour', 'time_of_day']].iloc[0:25])


# Analyze pedestrian activity by time of day (actual 24 hrs) using bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=df.groupby('hour')['Pedestrians'].mean().index,
            y=df.groupby('hour')['Pedestrians'].mean().values,
            palette='deep') #Using deep after checking sns.palettes.SEABORN_PALETTES.keys())
plt.xlabel("Time of Day")
plt.ylabel("Average Pedestrian Count")
plt.title("Pedestrian Activity by Time of Day")
plt.show()


# Now pedestrian activity by time of day classifications made using the function using a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=df.groupby('time_of_day')['Pedestrians'].mean().index,
            y=df.groupby('time_of_day')['Pedestrians'].mean().values,
            palette='pastel') #Using pastel after checking sns.palettes.SEABORN_PALETTES.keys())
plt.xlabel("Time of Day")
plt.ylabel("Average Pedestrian Count")
plt.title("Pedestrian Activity by Time of Day")
plt.show()


# Line plot instead of bar plot
plt.figure(figsize=(8, 5))
plt.plot(time_of_day_counts.index, time_of_day_counts.values, marker='s', color='r')

# Customizing the plot
plt.xlabel("Time of Day")
plt.ylabel("Average Pedestrian Count")
plt.title("Pedestrian Activity by Time of Day")
plt.show()

# Line plot instead of bar plot
plt.figure(figsize=(8, 5))
plt.plot(time_of_day_total.index, time_of_day_total.values, marker='s', color='r')

# Customizing the plot
plt.xlabel("Time of Day")
plt.ylabel("Total Pedestrian Count")
plt.title("Pedestrian Activity by Time of Day")
plt.show()
```
