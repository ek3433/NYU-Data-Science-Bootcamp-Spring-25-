````python
# Numpy Question 1: 1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.
import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

# Defining two 4x4 numpy arrays A and B (has to be 4x4 to do both the horizontal and vertical stacks)

A = iris_2d[17:21, :] 
B = iris_2d[47:51, :]  

print("Array A:\n", A,"\n Shape:", A.shape)
print("\nArray B:\n", B,"\n Shape:", B.shape)

# Vertical stack
v_stack = np.vstack((A, B))

# Horizontal stack
h_stack = np.hstack((A, B))

print("\nVertical Stack:\n", v_stack,"\n Shape:", v_stack.shape)
print("\nHorizontal Stack:\n", h_stack, "\n Shape:", h_stack.shape)
````

````python
#Numpy Question 2: Find common elements between A and B. [Hint : Intersection of two sets]

# Empty list that will be filled with common elements
common_elements = []

#Convert A and B into lists using flatten function
A_elements=A.flatten()
B_elements=B.flatten()

print("Elements of A:", A_elements)
print("Elements of B", B_elements)

# Iterating using for loops to find common elements
for a in A_elements:
    for b in B_elements:
        if (a == b) and (a not in common_elements):
            common_elements.append(a)

print("Common elements between A and B:", np.sort(common_elements))
````

````python
#Numpy Question 3: Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]
# Use np.where to find elements between 5 and 10
boolean_result = np.where((A >= 5) & (A <= 10))

# Extracting the values based on the indices returned by np.where
result = A[boolean_result]
print("Elements between 5 and 10 in array A are:", np.sort(result))
````

````python
#Numpy Question 4:Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

# Boolean conditions for filtering Petal length > 1.5 and Sepal length < 5.0
filtered_rows = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]

print("Filtered rows where petal length > 1.5 and sepal length < 5.0:")
print(filtered_rows)
print("Total number of rows:", len(iris_2d))
print("There are", len(filtered_rows), "that fit the conditions asked in the question.")
````

````python
#Pandas Question 2: Replace missing values in Min.Price and Max.Price columns with their respective mean.

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

#Calculate mean min and max prices
mean_minprice=df['Min.Price'].mean()
mean_maxprice=df['Max.Price'].mean()

#Display the obtained values:
print("Mean Minimum Price is: ",round(mean_minprice,1))
print("Mean Maximum Price is: ",round(mean_maxprice,1))

#The dataframe before missing (NaN) values are filled
display(df)

#Lambda functions to fill missing values for the respective columns (rounded to 1 digit after decimal)
min_apply = lambda x: round(mean_minprice, 1) if pd.isna(x) else round(x, 1)
max_apply = lambda x: round(mean_maxprice, 1) if pd.isna(x) else round(x, 1)

#Apply Lambda function of the respective columns
df['Min.Price'] = df['Min.Price'].apply(min_apply)
df['Max.Price'] = df['Max.Price'].apply(max_apply)

#Final dataframe with missing values filled
display(df)
````

````python
#Pandas Question 3: How to get the rows of a dataframe with row sum > 100?

df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))

#Add column to df to display the sum of the original rows
df['Sum'] = df.sum(axis=1)

display(df)

# List to store the rows with sum > 100
sum100 = []

# Loop through each row in the DataFrame
for i in range(0,len(df)):
    # Calculate the sum of the rows (excluding the cell corresponding to the newly added Sum column)
    row_sum = sum(df.loc[i,0:3])
    
    # If the row sum is greater than 100, add the row to the list
    if row_sum > 100:
        sum100.append(i)

# Convert the list of rows back to a DataFrame
result_df = pd.DataFrame(sum100)
result_df.rename(columns={0: 'row'}, inplace=True)

# Display the result DataFrame 
# Note: This df will display the index values starting from 0, if we want to assume rows start from 1, the code can be changed to append (i+1)
display(result_df)
````
